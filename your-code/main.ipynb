```python
# your code here
from scipy.stats import uniform

# Set parameters for the uniform distribution
low = 4
high = 20   
size = 10   

data= uniform.rvs(4, 20, size=100)

# Plot histogram
plt.hist(data, bins=20, color='skyblue', edgecolor='black', density=True)
plt.title("Uniform Distribution between {} and {}".format(low, high))
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from scipy.stats import norm 
```

```python
# your code here
# Define the mean and standard deviation for normal distribution
mean = 10 # Mean 
std_deviation = 1  # Standard deviation 

# Generate synthetic data following a normal distribution
data = 1000  # Number of employees 
norm_distribution1= np.random.normal(mean, std_deviation, data)
```

```python
# your code here
# Plot a histogram of normal distribution
plt.figure(figsize=(8, 6))
plt.hist(norm_distribution1, bins=70, density=True, alpha=0.6, color='b', label='Data')

# Generate a range of x values for the probability density function (PDF)
x_values = np.linspace(min(norm_distribution1), max(norm_distribution1), 1000)

# Calculate the PDF using the scipy.stats.norm library
pdf_values = norm.pdf(x_values, mean, std_deviation)

# Plot the PDF as a smooth curve
plt.plot(x_values, pdf_values, 'r-', lw=2, label='Normal Distribution PDF')

# Add labels and a legend to the plot
plt.xlabel('Salary (USD)')
plt.ylabel('Probability Density')
plt.title('Employee Salary Distribution')
plt.legend()

# Show the plot
plt.show()
```

```python
# Define the mean and standard deviation
mean = 10              # Mean (center) of the distribution
std_deviation = 50      # Standard deviation (spread) of the distribution
# Define the number of samples to generate
data = 1000            # Number of data points to generate
# Generate synthetic data following a normal (Gaussian) distribution
# np.random.normal(loc, scale, size)
#   loc   = mean
#   scale = standard deviation
#   size  = number of values to generate
norm_distribution = np.random.normal(mean, std_deviation, data)
# Display the first 10 generated values (optional)
print(norm_distribution[:10])
```

```python
# Plot a histogram of generated data
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(norm_distribution, bins=70, density=True, alpha=0.6, color='b', label='Generated Data')
x_values = np.linspace(min(norm_distribution), max(norm_distribution), 1000)
# Calculate the corresponding PDF (Probability Density Function) values
# norm.pdf(x, mean, std_dev) gives the height of the normal curve for each x
pdf_values = norm.pdf(x_values, mean, std_deviation)
# Plot the PDF as a smooth red curve
# 'r-' means red line, lw=2 means line width of 2
plt.plot(x_values, pdf_values, 'r-', lw=2, label='Normal Distribution PDF')
# Add descriptive labels and legend
plt.xlabel('Value')                    # Label for the x-axis
plt.ylabel('Probability Density')      # Label for the y-axis
plt.title('Normal Distribution (mean=10, std=1)')  # Plot title
plt.legend()                           # Show legend box
# Display the plot
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

How are the two distributions different?

```python
 #my mean is equal to the scale of the graph

def exo_dist(mean, size):
    data = np.random.exponential(mean, size)
    plt.figure(figsize=(10, 5))
    plt.hist(data, bins=100, edgecolor='black', color='skyblue')
    plt.title(f'Exponential Distribution: scale={mean}, size={size}')
    plt.xlabel('Value')
    plt.ylabel('Count')
    plt.grid(True)
```

```python
exo_dist(10,10)
```

```python
exo_dist(10,100)
```

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
wahrscheinlichkeit = 0
for k in range (1,15):
    pos = ((1/10)**k*math.exp(1/10))/math.factorial(k)
    wahrscheinlichkeit +=pos
print(wahrscheinlichkeit)
```

```python
# your answer here
# Hint: This is same as saying P(x<15)
```

```python
# your code here
# Probabilty that the customer will spend more than 15 minutes
1 - wahrscheinlichkeit 
```
